```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import SparsePauliOp
from qiskit_ibm_runtime import Estimator as Estimator

bell = QuantumCircuit(3,3)
bell.x(1)
bell.h(0)
bell.cx(0,2)
bell.barrier()
bell.measure(0,0)
bell.measure(1,1)
bell.measure(2,2)
bell.draw("mpl")
```
